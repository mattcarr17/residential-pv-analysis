### Data Cleaning

```python
solar.isna().sum()
```

-1 used to indicate missing values

```python
for col in COLS:
    num_missing = len(solar[solar[col] == -1])
    print(col + ': ', num_missing)
```

Only about 8,000 missing `system_size_DC`. Will investigate these rows more closely in the future to check for any bias in removing them, for now I drop them

```python
solar = solar[solar['system_size_DC'] != -1]
solar.shape
```

Still nearly 2 million PV systems to analyze. `zip_code` missing over 100,000 entries. Explore the states missing `zip_code` entries

```python
for state in solar['state'].unique():
    systems = solar[solar['state'] == state]
    total_systems = len(systems)
    missing_zip = len(systems[systems['zip_code'] == -1])
    percent_missing = round(missing_zip/total_systems, 2) * 100

    print(f'{state}: {percent_missing}%')
```

Colorado and Rhode Island systems are all missing `zip_code` values. 88% of Maryland and 51% of Minnesota. Dropping all PV systems missing `zip_code`

```python
solar = solar[solar['zip_code'] != -1]
```

```python
solar['new_construction'].value_counts()
```

Need to remove four digit extension

```python
solar['zcta'] = solar['zip_code'].astype(str).apply(lambda x: x.split('-')[0])
```

```python
solar['zcta'].value_counts()
```

Drop rows from Maryland and Minnesota. There are a lot of rows from those states where the ZCTA is missing, but they still have installations. If I leave the states in, all the rows missing the ZCTA value will be considered to have no PV installations, which would be incorrect. I'd rather have less data than rows I know to be incorrect. A few other states have very few missing rows so I am going to leave those (Arizona, California). I might be able to check the previous version of the data to see if it is missing there. For now I will be using this data alone.

```python
solar = solar.loc[(solar['state'] != 'MD') & (solar['state'] != 'MN')]
```

```python
solar.groupby('state').agg('nunique')['zcta']
```

```python
solar['zcta'].value_counts().sort_index()
```

About 4,400 ZCTAs are listed as an empty string. This might be from how I calculated the `zcta` column

```python
solar.loc[solar['zip_code'] == '']
```

```python
solar.loc[solar['zcta'] == '']['zip_code']
```

All `zcta` with an empty string are missing. Drop these rows.

```python
solar = solar.loc[solar['zcta'] != '']
```

```python
solar.loc[solar['zcta'] == '0']['zip_code']
```

```python
solar = solar.loc[solar['zcta'] != '0']
```

```python
solar['zcta'].value_counts().sort_index()
```

All `zcta` values are valid now for the most part. In the future I will need to go through and do a more robust prep of the data but for the MVP this will work.

Finally I can aggregate the data by `zcta`. I want the following info for each `zcta`:
- total number of PV systems
- total kW (sum `system_size_DC`)
- average system size
- median system size

```python
pv_data = solar.groupby('zcta').agg(['count', 'sum', 'mean', 'median'])['system_size_DC']
pv_data.head()
```

```python
pv_data = pv_data.reset_index()
pv_data.head()
```

```python
pv_data.columns = ['zcta', 'num_systems', 'total_capacity', 'mean_system_size', 'median_system_size']
pv_data.head()
```

Save as a csv to be used in future notebooks

```python
pv_data.to_csv('../data/pv_systems.csv')
```
